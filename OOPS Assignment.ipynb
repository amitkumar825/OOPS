```python
#1. What are the five key concepts of Object-Oriented Programming (OOP)?

Answer:-

#1.Encapsulation:- The idea of wrapping data and the methods that work on it within a single unit. This limits direct access to variables and methods, which can prevent accidental data modification
Ex- 
class student:
    def __init__(self, name, degree):
        self.name = name
        self.degree = degree
stud1 = student("amit", "b.tech")
stud1.name
Results- 'amit'
stud1.degree
Results-'b.tech'



#2.Inheritance:- Allows attributes and methods to be acquired from a base class to a derived class.
Ex- 
class father:
    def father_property(self):
        print("this is the father property")
        
class son(father):
    def job(self):
        print("son has property from job")
child_obj = son()
child_obj.job()
Results- son has property from job
child_obj.father_property()
Results- this is the father property
father_obj = father()
father_obj.father_property()
Results- this is the father property
        
        

#3.Polymorphism:- Promotes the dynamic inheritance of child classes from their parent class.
Ex-
class student:
    def student(self):
        print("welcome to pwskills class")
    def student(self, name = ""):
        print("welcome to pwskills class", name)
    def student(self, name = "", course = ""):
        print("welcome to pwskills class", name, course)

stud = student()
stud.student()
Results- welcome to pwskills class       
stud.student("amit")
Results- welcome to pwskills class amit
stud.student('amit', "ds")
Results- welcome to pwskills class amit ds
        
    
        
#4.Abstraction:- Involves abstracting certain details of an object's complexity to present a concrete model.
Ex-
import abc
class shape:
    
    @abc.abstractmethod
    def calculate_area(self):
        pass
    
class rectangle(shape):
    def calculate_area(self):
        return "area of rectange is len*breadth"
    
class circle(shape):
    def calculate_area(self):
        return "area of circle is pi r **2"
    
rect = rectangle()
rect.calculate_area()
Results- 'area of rectange is len*breadth'



#5. Classes:- Classes are blueprints for creating objects. They define a type of object by bundling data and methods that operate on the data.
Ex-
class student:
    def __init__(self, name):
        self.name = name
obj = student("amit")
obj.name
Results- 'amit'
```

```python
#2. Write a Python class for a 'Car` with attributes for 'make', 'model', and 'year'. Include a method to display the car's information.

Answer:-

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

if __name__ == "__main__":
    my_car = Car("Toyota", "Camry", 2022)

print(my_car.display_info())
Results- 2022 Toyota Camry
```

```python
#3. Explain the difference between instance methods and class methods. Provide an example of each.

Answer:-

#Instance methods:- Instance methods operate on an instance of the class. They can access and modify the attributes of that instance and are called on an instance of the class.
Ex-
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

    def calculate_perimeter(self):
        return 2 * (self.width + self.height)
    
rect = Rectangle(5, 10)
print(rect.calculate_area())
Results- 50
print(rect.calculate_perimeter())
Results- 30



#Class methods:- Class methods operate on the class itself rather than on instances of the class. They can access and modify class-level attributes and are called on the class rather than on an instance.
Ex- 
class student:
    def __init__(self, name):
        self.name = name 
    
    @classmethod 
    def student_details(cls, name): 
        return cls(name)
    
obj1 = student.student_details("amit")
obj1.name
Results- 'amit'
```

```python
#4. How does Python implement method overloading? Give an example.

Answer:-

class Calculator:
    def add(self, a, b=None):
        if b is None:
            return lambda x: a + x
        else:
            return a + b
        
calc = Calculator()
print(calc.add(10, 13))
Results- 23
calc = calc.add(12)
print(calc(10))
Results- 22
```

```python
#6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.

Answer:-

#1.single inheritance:- when a derived class has one parent class
Ex-
class fruit:
    def fruit_info(self):
        print("inside parent class")
class apple(fruit):
    def apple_info(self):
            print("inside the child class")
            
obj = apple()
obj.apple_info()
Results- inside the child class
obj.fruit_info()
Results- inside parent class
par = fruit()
par.fruit_info()
Results- inside parent class


#2.Multi-level inheritance
Ex-
class vehicle:
    def  vehicle_info(self):
        print("inside vehicle class")
class car(vehicle):
    def car_info(self):
        print("inside the car class")
class sportscar(car):
    def sports_car_info(self):
        print("inside sports car")
        
s1 = sportscar()
s1.sports_car_info()
Results-inside sports car
s1.vehicle_info()
Results- inside vehicle class
s1.car_info()
Results- inside the car class


#3. multiple inheritance:- one child class may inherit the property of multiple parents
Ex-
class parentclass1:
    def method1(self):
        print("method 1 of parent class 1")
class parentclass2:
    def method2(self):
        print("method 2 of parent calss 2")
        
class childclass(parentclass1, parentclass2):
    def method(self):
        print("method of child class")
        
        
c1 = childclass()
c1.method()
Results- method of child class
c1.method1()
Results- method 1 of parent class 1
c1.method2()
Results- method 2 of parent calss 2


#4.hierarchical inheritance:- one parent class>>multiple(more than two childclass)
Ex-
class vehicle:
    def info(self):
        print("this is vehicle")
class car(vehicle):
    def car_info(self, name):
        print("this is car info", name)
class truck(vehicle):
    def truck_info(self, name):
        print("truck info", name)
        
c1 = car()
c1.info()
Results- this is vehicle
c1.car_info("BMW")
Results- this is car info BMW
obj2 = truck()
obj2.info()
Results- this is vehicle
obj2.truck_info("truck1")
Results- truck info truck1


#5. hybrid inheritance:- multiple types or combination of different inheritance
Ex-
class vehicle:
    def vehicle_info(self):
        print("inside the vehicle class")

class car(vehicle):
    def car_info(self):
        print("inside the car class")
    
class truck(vehicle):
    def truck_info(self, name):
        print("inside truck info", name)
        
class sportscar(car, vehicle):
    def sports_info(self):
        print("inside sports car class")
        
obj1 = sportscar()
obj1.car_info()
Results- inside the car class
obj1.vehicle_info()
Results- inside the vehicle class
obj1.car_info()
inside the car class
```

```python
#7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

Answer:-

#Method Resolution Order (MRO):- the Method Resolution Order (MRO) is a sequence that determines the order in which base classes are looked up when searching for a method. This is particularly important in the context of multiple inheritance, where a class can inherit from more than one base class.

#Retrieve it programmatically
Ex-
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

print(D.__mro__)
Results- (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
print(D.mro())
Results- [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

```python
#8. Create an abstract base class 'Shape` with an abstract method area(). Then create two subclasses 'Circle` and Rectangle that implement the 'area() method.

Answer:-

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        """Method to calculate the area of the shape"""
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        """Calculate the area of the circle"""
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        """Calculate the area of the rectangle"""
        return self.width * self.height

if __name__ == "__main__":
    circle = Circle(5)
    rectangle = Rectangle(4, 6)
    
print(f"Circle area: {circle.area()}")
Results- Circle area: 78.53981633974483
print(f"Rectangle area: {rectangle.area()}") 
Results- Rectangle area: 24
```

```python
#10. Implement encapsulation in a 'BankAccount` class with private attributes for 'balance` and 'account_number`. Include methods for deposit, withdrawal, and balance inquiry.

Answer:-

class BankAccount:
    def __init__(self, account_number, initial_balance=0):
       
        self.__account_number = account_number
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew {amount}. New balance: {self.__balance}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        return self.__balance

    def get_account_number(self):
        return self.__account_number
    
account = BankAccount("1234567890", 5000)
print(f"Account Number: {account.get_account_number()}")
Results- Account Number: 1234567890
account.deposit(100)
Results- Deposited 100. New balance: 5100.
account.withdraw(1000)
Results- Withdrew 1000. New balance: 4100.
print(f"Final Balance: {account.get_balance()}")
Results- Final Balance: 4100
```

```python
#11. Write a class that overrides the_str` and `add_magic methods. What will these methods allow you to do?

Answer:-

class CustomNumber:
    def __init__(self, value):
        self.value = value
    def __str__(self):
        return f"CustomNumber({self.value})"
    def __add__(self, other):
        if isinstance(other, CustomNumber):
            return CustomNumber(self.value + other.value)
        return NotImplemented
    
a = CustomNumber(87)
b = CustomNumber(56)
print(a) 
Results- CustomNumber(87)
print(b) 
Results- CustomNumber(56)
c = a + b
print(c) 
Results- CustomNumber(143)
```

```python
#12. Create a decorator that measures and prints the execution time of a function.

Answer:-

import time
def timer_decorator(func):
    def timer():
        start = time.time()
        func()
        end = time.time()
        print("the time for executing the code", end-start)
    return timer

@timer_decorator
def func_test():
    print(11*1000)
func_test()
Results-
11000
the time for executing the code 5.888938903808594e-05
```

```python
#14. Write a class method that keeps track of the number of instances created from a class.

Answer:-

class InstanceTracker:
    _instance_count = 0

    def __init__(self):
        InstanceTracker._instance_count += 1

    @classmethod
    def get_instance_count(cls):
        return cls._instance_count
    
obj1 = InstanceTracker()
print(InstanceTracker.get_instance_count())
Results- 1

obj2 = InstanceTracker()
print(InstanceTracker.get_instance_count())
Results- 2
```

```python
#15. Implement a static method in a class that checks if a given year is a leap year.

Answer:-

class YearUtils:
    @staticmethod
    def is_leap_year(year):
        """
        Check if a given year is a leap year.

        Parameters:
        year (int): The year to check.

        Returns:
        bool: True if the year is a leap year, False otherwise.
        """
        if year % 400 == 0:
            return True
        if year % 100 == 0:
            return False
        if year % 4 == 0:
            return True
        return False
    
year = 2024
print(f"Is {year} a leap year? {YearUtils.is_leap_year(year)}")
Results- Is 2024 a leap year? True
```
